#### Insertion of a Key into Trie

```python
class TrieNode:
    def __init__(self):
        # Initialize links array and isEnd flag
        self.links = [None] * 26
        self.is_end = False

    def contains_key(self, ch: str) -> bool:
        return self.links[ord(ch) - ord('a')] is not None

    def get(self, ch: str) -> 'TrieNode':
        return self.links[ord(ch) - ord('a')]

    def put(self, ch: str, node: 'TrieNode') -> None:
        self.links[ord(ch) - ord('a')] = node

    def set_end(self) -> None:
        self.is_end = True

    def is_end(self) -> bool:
        return self.is_end
```

Insertion of a Key into a Trie

To insert a key into a Trie, follow these steps:

    Start at the root and search for a link corresponding to the first character of the key.
        If the link exists, move to the child node and repeat for the next character.
        If the link does not exist, create a new node and link it to the parent node.
    Continue until you reach the last character of the key, then mark the final node as an end node.


```python
class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        node = self.root #a Trie
        for ch in word:
            if not node.contains_key(ch):
                node.put(ch, TrieNode())
            node = node.get(ch)
        node.set_end()
```
