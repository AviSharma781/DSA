#### The Power Sum

```python
# https://www.hackerrank.com/challenges/the-power-sum/problem?isFullScreen=true
```

```python
import math
X = 64
x_ = math.floor(np.sqrt(X))
ans = []
input_lst = list(range(1, x_+1))
power = 2

def backtrack(curr, i):
    
    sum_ = sum([k**power for k in curr]) 
    
    if sum_ == X:
        ans.append(curr[:])
    
    if i==len(input_lst):
           return 
    
    for j in range(i, len(input_lst)):
        curr.append(input_lst[j])
        backtrack(curr, j+1)
        curr.pop()

backtrack([], 0)
ans  
```
